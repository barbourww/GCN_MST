```python
import barbour_lavanne_proj2_class as proj
import itertools
```

```python
n_nodes = 40
n_graphs = 200
tf = 0.5
vf = 0.25

layers = [('gc', n_nodes, 'B'), ('gc', n_nodes*8, 'B'), ('d', n_nodes, ''), ('d', n_nodes*8, '')]
features = ['degree', 'weighted_degree', 'weighted_second_degree', 'path']

# make list of features of length >= 2
feat = list(itertools.chain.from_iterable([list(itertools.combinations(features, l)) for l in range(2, len(features)+1)]))
# add single feature for reference
feat += [('weighted_degree',)]
# make list of layer configurations of length >= 1, <= 4
# itertools.product used for repetition of layers
lyrcfs = list(itertools.chain.from_iterable([itertools.product(layers, repeat=l) for l in [1, 2, 3, 4]]))
print(len(lyrcfs))
# add last layer (regression pooling) to each configuration
lyrcfs = [lc + (('d', 1, ''),) for lc in lyrcfs]

# make all possible trial combinations
p = list(itertools.product(feat, lyrcfs))
print("Total trials:", len(p))

m = proj.MST_GCN(log=True)

# put features in product first so that they won't regenerate each time 
# (i.e. itertools.product loops through second items for each first item)
i = 0
for f, lc in p:
    i += 1
    print(i, '---', f, lc)
    m.train_model(n_graphs=n_graphs, n_nodes=n_nodes, train_frac=tf, val_frac=vf, feature_gen=f, layer_config=lc, dropouts=[False]*len(lc))
```
