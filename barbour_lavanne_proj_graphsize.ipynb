```python
import barbour_lavanne_proj2_class as proj
```

```python
n_graphs = 200
tf = 0.5
vf = 0.25
feat = ['weighted_degree', 'degree']
m = proj.MST_GCN(log=True)
```

```python
m.train_model(n_graphs=n_graphs, n_nodes=20, train_frac=tf, val_frac=vf, 
              feature_gen=feat, force_regen_data=True,
              layer_config=[('gc', 20, 'AB'), ('d', 1, '')], dropouts=[False, False])
m.plot_results()
m.train_model(n_graphs=n_graphs, n_nodes=40, train_frac=tf, val_frac=vf, 
              feature_gen=feat, force_regen_data=True,
              layer_config=[('gc', 40, 'AB'), ('d', 1, '')], dropouts=[False, False])
m.plot_results()
m.train_model(n_graphs=n_graphs, n_nodes=60, train_frac=tf, val_frac=vf, 
              feature_gen=feat, force_regen_data=True,
              layer_config=[('gc', 60, 'AB'), ('d', 1, '')], dropouts=[False, False])
m.plot_results()
m.train_model(n_graphs=n_graphs, n_nodes=80, train_frac=tf, val_frac=vf, 
              feature_gen=feat, force_regen_data=True,
              layer_config=[('gc', 80, 'AB'), ('d', 1, '')], dropouts=[False, False])
m.plot_results()
m.train_model(n_graphs=n_graphs, n_nodes=100, train_frac=tf, val_frac=vf, 
              feature_gen=feat, force_regen_data=True,
              layer_config=[('gc', 100, 'AB'), ('d', 1, '')], dropouts=[False, False])
m.plot_results()
```

```python
n_nodes = 100
m.train_model(n_graphs=n_graphs, n_nodes=n_nodes, train_frac=0.5, val_frac=0.3, 
              feature_gen=['weighted_degree'],
              layer_config=[('gc', n_nodes, 'B'), ('d', 1, '')], 
              dropouts=[False, False], test_n_nodes=int(n_nodes*0.4), verbose=False)
m.plot_results()
m.train_model(n_graphs=n_graphs, n_nodes=n_nodes, train_frac=0.5, val_frac=0.3, 
              feature_gen=['weighted_degree'],
              layer_config=[('gc', n_nodes, 'B'), ('d', 1, '')], 
              dropouts=[False, False], test_n_nodes=int(n_nodes*0.6), verbose=False)
m.plot_results()
m.train_model(n_graphs=n_graphs, n_nodes=n_nodes, train_frac=0.5, val_frac=0.3, 
              feature_gen=['weighted_degree'],
              layer_config=[('gc', n_nodes, 'B'), ('d', 1, '')], 
              dropouts=[False, False], test_n_nodes=int(n_nodes*0.8), verbose=False)
m.plot_results()
m.train_model(n_graphs=n_graphs, n_nodes=n_nodes, train_frac=0.5, val_frac=0.3, 
              feature_gen=['weighted_degree'],
              layer_config=[('gc', n_nodes, 'B'), ('d', 1, '')], 
              dropouts=[False, False], test_n_nodes=None, verbose=False)
m.plot_results()
```
