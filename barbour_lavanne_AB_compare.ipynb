```python
import barbour_lavanne_proj2_class as proj
```

# Comparison of NN structure

```python
n_nodes = 50
n_graphs = 100
tf = 0.5
vf = 0.25
feat = ['weighted_degree']
m = proj.MST_GCN(log=True)
```

```python
m.train_model(n_graphs=n_graphs, n_nodes=n_nodes, train_frac=tf, val_frac=vf, 
              feature_gen=feat,
              layer_config=[('gc', n_nodes, 'B'), ('d', 1, '')], dropouts=[False, False])
m.plot_results()
m.train_model(n_graphs=n_graphs, n_nodes=n_nodes, train_frac=tf, val_frac=vf, 
              feature_gen=feat,
              layer_config=[('gc', n_nodes*10, 'B'), ('d', 1, '')], dropouts=[False, False])
m.plot_results()
m.train_model(n_graphs=n_graphs, n_nodes=n_nodes, train_frac=tf, val_frac=vf, 
              feature_gen=feat,
              layer_config=[('gc', n_nodes, 'B'), ('d', n_nodes, ''), ('d', 1, '')], dropouts=[False, False, False])
m.plot_results()
m.train_model(n_graphs=n_graphs, n_nodes=n_nodes, train_frac=tf, val_frac=vf, 
              feature_gen=feat,
              layer_config=[('gc', n_nodes*5, 'B'), ('d', n_nodes*5, ''), ('d', 1, '')], dropouts=[False, False, False])
m.plot_results()
m.train_model(n_graphs=n_graphs, n_nodes=n_nodes, train_frac=tf, val_frac=vf, 
              feature_gen=feat,
              layer_config=[('gc', n_nodes, 'B'), ('d', n_nodes, ''), ('d', n_nodes, ''), ('d', 1, '')], dropouts=[False, False, False, False])
m.plot_results()
m.train_model(n_graphs=n_graphs, n_nodes=n_nodes, train_frac=tf, val_frac=vf, 
              feature_gen=feat,
              layer_config=[('gc', n_nodes, 'B'), ('gc', n_nodes, 'B'), ('d', n_nodes, ''), ('d', 1, '')], dropouts=[False, False, False, False])
m.plot_results()
```

# Comparison of input graph connectivity
